```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
train_data = pd.read_csv("train.csv")
```

```python
device = train_data['device'].apply(json.loads)
device_list = list(device)
device_json = json.dumps(device_list)
train_data = train_data.join(pd.read_json(device_json))
del(train_data['device'])
```

```python
device = train_data['geoNetwork'].apply(json.loads)
device_list = list(device)
device_json = json.dumps(device_list)
train_data = train_data.join(pd.read_json(device_json))
del(train_data['geoNetwork'])
```

```python
device = train_data['totals'].apply(json.loads)
device_list = list(device)
device_json = json.dumps(device_list)
train_data = train_data.join(pd.read_json(device_json))
del(train_data['totals'])
```

```python
device = train_data['trafficSource'].apply(json.loads)
device_list = list(device)
device_json = json.dumps(device_list)
train_data = train_data.join(pd.read_json(device_json))
del(train_data['trafficSource'])
```

```python
constant_column = []
for col in train_data.columns:
    if len(train_data[col].value_counts())==1:
        constant_column.append(col)
```
